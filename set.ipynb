# Sets em Python 🐍

## O que é um Set?

Um **Set** (conjunto) é uma estrutura de dados utilizada para armazenar uma coleção de elementos **únicos**, ou seja, **não permite valores duplicados**.

Os sets são muito úteis quando precisamos:

* Eliminar valores repetidos;
* Verificar rapidamente se um elemento existe;
* Realizar operações matemáticas de conjuntos (união, interseção e diferença).

---

## Características dos Sets

✔ Não permitem elementos duplicados

✔ Não possuem índices

✔ Não garantem a ordem dos elementos

✔ São mutáveis (podem ser alterados)

✔ Utilizam chaves `{}`

---

## Criando um Set

```python
frutas = {"maçã", "banana", "uva"}

print(frutas)
```

Saída:

```python
{'banana', 'uva', 'maçã'}
```

Observe que a ordem pode ser diferente da inserida.

---

## Removendo Valores Duplicados

Uma das aplicações mais comuns dos sets é remover elementos repetidos.

```python
numeros = [1, 2, 2, 3, 3, 3, 4, 5]

unicos = set(numeros)

print(unicos)
```

Saída:

```python
{1, 2, 3, 4, 5}
```

---

## Adicionando Elementos

Utilizamos o método `add()`.

```python
cores = {"azul", "verde"}

cores.add("vermelho")

print(cores)
```

Saída:

```python
{'azul', 'verde', 'vermelho'}
```

---

## Tentando Adicionar um Elemento Duplicado

```python
cores = {"azul", "verde"}

cores.add("azul")

print(cores)
```

Saída:

```python
{'azul', 'verde'}
```

Nada acontece porque o valor já existe.

---

## Removendo Elementos

### remove()

```python
animais = {"cachorro", "gato", "coelho"}

animais.remove("gato")

print(animais)
```

Saída:

```python
{'cachorro', 'coelho'}
```

---

### discard()

Diferente do `remove()`, não gera erro caso o elemento não exista.

```python
animais = {"cachorro", "gato"}

animais.discard("leão")

print(animais)
```

Saída:

```python
{'cachorro', 'gato'}
```

---

## Verificando se um Elemento Existe

Utilizamos o operador `in`.

```python
frutas = {"maçã", "banana", "uva"}

print("banana" in frutas)
print("laranja" in frutas)
```

Saída:

```python
True
False
```

---

## Percorrendo um Set

```python
alunos = {"Ana", "Carlos", "Bruno"}

for aluno in alunos:
    print(aluno)
```

A ordem pode variar a cada execução.

# Operações entre Sets

## União

Junta todos os elementos dos conjuntos.

```python
a = {1, 2, 3}
b = {3, 4, 5}

resultado = a.union(b)

print(resultado)
```

Saída:

```python
{1, 2, 3, 4, 5}
```

---

## Interseção

Retorna apenas os elementos em comum.

```python
a = {1, 2, 3}
b = {3, 4, 5}

resultado = a.intersection(b)

print(resultado)
```

Saída:

```python
{3}
```

---

## Diferença

Retorna os elementos que existem em um conjunto mas não no outro.

```python
a = {1, 2, 3}
b = {3, 4, 5}

resultado = a.difference(b)

print(resultado)
```

Saída:

```python
{1, 2}
```

---

## Diferença Simétrica

Retorna elementos que estão em apenas um dos conjuntos.

```python
a = {1, 2, 3}
b = {3, 4, 5}

resultado = a.symmetric_difference(b)

print(resultado)
```

Saída:

```python
{1, 2, 4, 5}
```

# Tamanho do Set

Utilizamos a função `len()`.

```python
frutas = {"maçã", "banana", "uva"}

print(len(frutas))
```

Saída:

# Exemplo Prático 1 – Controle de Presença

```python
presentes = {"Ana", "Bruno", "Carlos"}

novo_aluno = input("Digite o nome do aluno: ")

if novo_aluno in presentes:
    print("Aluno já registrado.")
else:
    presentes.add(novo_aluno)
    print("Aluno adicionado.")

print(presentes)
```

---

# Exemplo Prático 2 – Removendo Duplicados de uma Lista

```python
nomes = [
    "Ana",
    "Bruno",
    "Ana",
    "Carlos",
    "Bruno",
    "Maria"
]

nomes_unicos = list(set(nomes))

print(nomes_unicos)
```
