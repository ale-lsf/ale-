# Dicionários em Python 🐍

## O que é um Dicionário?

Um **dicionário (dictionary)** é uma estrutura de dados que armazena informações no formato **chave : valor**.

Podemos imaginar um dicionário como uma ficha cadastral:

| Chave  | Valor    |
| ------ | -------- |
| nome   | Ana      |
| idade  | 20       |
| cidade | Blumenau |

Em Python, os dicionários são criados utilizando `{}`.

---

# Características dos Dicionários

✔ Armazenam dados em pares **chave:valor**

✔ As chaves devem ser únicas

✔ Permitem diferentes tipos de dados

✔ São mutáveis (podem ser alterados)

✔ Mantêm a ordem de inserção (Python 3.7+)

---

# Criando um Dicionário

```python
aluno = {
    "nome": "Ana",
    "idade": 20,
    "cidade": "Blumenau"
}

print(aluno)
```

Saída:

```python
{'nome': 'Ana', 'idade': 20, 'cidade': 'Blumenau'}
```

---

# Acessando Valores

Para acessar um valor, utilizamos a chave.

```python
aluno = {
    "nome": "Ana",
    "idade": 20
}

print(aluno["nome"])
```

Saída:

```python
Ana
```

---

# Adicionando Novos Elementos

```python
aluno = {
    "nome": "Ana",
    "idade": 20
}

aluno["curso"] = "Desenvolvimento de Sistemas"

print(aluno)
```

Saída:

```python
{
 'nome': 'Ana',
 'idade': 20,
 'curso': 'Desenvolvimento de Sistemas'
}
```

---

# Alterando Valores

```python
aluno = {
    "nome": "Ana",
    "idade": 20
}

aluno["idade"] = 21

print(aluno)
```

Saída:

```python
{'nome': 'Ana', 'idade': 21}
```

---

# Removendo Elementos

## Utilizando `del`

```python
aluno = {
    "nome": "Ana",
    "idade": 20,
    "cidade": "Blumenau"
}

del aluno["cidade"]

print(aluno)
```

Saída:

```python
{'nome': 'Ana', 'idade': 20}
```

---

## Utilizando `pop()`

```python
aluno = {
    "nome": "Ana",
    "idade": 20
}

idade = aluno.pop("idade")

print(idade)
print(aluno)
```

Saída:

```python
20
{'nome': 'Ana'}
```

---

# Verificando se uma Chave Existe

```python
aluno = {
    "nome": "Ana",
    "idade": 20
}

print("nome" in aluno)
print("curso" in aluno)
```

Saída:

```python
True
False
```

---

# Percorrendo um Dicionário

## Apenas as Chaves

```python
aluno = {
    "nome": "Ana",
    "idade": 20,
    "cidade": "Blumenau"
}

for chave in aluno:
    print(chave)
```

Saída:

```python
nome
idade
cidade
```

---

## Apenas os Valores

```python
for valor in aluno.values():
    print(valor)
```

Saída:

```python
Ana
20
Blumenau
```

---

## Chaves e Valores

```python
for chave, valor in aluno.items():
    print(chave, "=", valor)
```

Saída:

```python
nome = Ana
idade = 20
cidade = Blumenau
```

# Método `get()`

O método `get()` evita erros caso a chave não exista.

### Exemplo com erro

```python
aluno = {
    "nome": "Ana"
}

print(aluno["idade"])
```

Saída:

```python
KeyError
```

---

### Utilizando `get()`

```python
aluno = {
    "nome": "Ana"
}

print(aluno.get("idade"))
```

Saída:

```python
None
```

---

### Valor padrão

```python
print(aluno.get("idade", "Não informado"))
```

Saída:

```python
Não informado
```

# Lista de Dicionários

Muito utilizada em sistemas e bancos de dados.

```python
alunos = [
    {"nome": "Ana", "idade": 20},
    {"nome": "Carlos", "idade": 21},
    {"nome": "Maria", "idade": 19}
]

for aluno in alunos:
    print(aluno["nome"])
```

Saída:

```python
Ana
Carlos
Maria
```

# Exemplo Prático – Agenda Telefônica

```python
agenda = {
    "Ana": "99999-1111",
    "Carlos": "99999-2222",
    "Maria": "99999-3333"
}

nome = input("Digite o nome: ")

if nome in agenda:
    print("Telefone:", agenda[nome])
else:
    print("Contato não encontrado.")
```

# Comparação: Lista x Dicionário

| Lista               | Dicionário           |
| ------------------- | -------------------- |
| Acesso por índice   | Acesso por chave     |
| `[0]`, `[1]`, `[2]` | `"nome"`, `"idade"`  |
| Mantém sequência    | Organiza por chave   |
| Ideal para coleções | Ideal para cadastros |

### Exemplo

**Lista**

```python
aluno = ["Ana", 20, "Blumenau"]
```

Difícil saber o que cada posição representa.

**Dicionário**

```python
aluno = {
    "nome": "Ana",
    "idade": 20,
    "cidade": "Blumenau"
}
```

Mais organizado e legível.

---

# Resumo

| Operação             | Comando                            |
| -------------------- | ---------------------------------- |
| Criar                | `{}`                               |
| Acessar valor        | `dicionario["chave"]`              |
| Acessar seguro       | `get()`                            |
| Adicionar            | `dicionario["nova_chave"] = valor` |
| Alterar              | `dicionario["chave"] = novo_valor` |
| Remover              | `del` ou `pop()`                   |
| Chaves               | `keys()`                           |
| Valores              | `values()`                         |
| Chaves e valores     | `items()`                          |
| Verificar existência | `in`                               |

### Quando usar Dicionários?

✅ Cadastros de alunos

✅ Produtos

✅ Usuários de sistemas

✅ Configurações de aplicações

✅ Dados vindos de APIs (JSON)

✅ Informações organizadas por atributos (nome, idade, cidade, etc.)
