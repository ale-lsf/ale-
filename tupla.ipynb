# Tuplas em Python 🐍

## O que é uma Tupla?

Uma **tupla (tuple)** é uma coleção ordenada de elementos que, depois de criada, **não pode ser alterada**.

Isso significa que os valores armazenados em uma tupla são **imutáveis**.

As tuplas são definidas utilizando **parênteses `()`**.

---

# Características das Tuplas

✔ Mantêm a ordem dos elementos

✔ Permitem valores repetidos

✔ Permitem diferentes tipos de dados

✔ São imutáveis

✔ Utilizam parênteses `()`

---

# Criando uma Tupla

```python
frutas = ("maçã", "banana", "uva")

print(frutas)
```

Saída:

```python
('maçã', 'banana', 'uva')
```

---

# Acessando Elementos

Como as tuplas são ordenadas, cada elemento possui um índice.

```python
frutas = ("maçã", "banana", "uva")

print(frutas[0])
print(frutas[1])
print(frutas[2])
```

Saída:

```python
maçã
banana
uva
```

---

# Índices Negativos

Os índices negativos começam pelo final da coleção.

```python
frutas = ("maçã", "banana", "uva")

print(frutas[-1])
print(frutas[-2])
```

Saída:

```python
uva
banana
```

---

# Tamanho da Tupla

```python
frutas = ("maçã", "banana", "uva")

print(len(frutas))
```

Saída:

```python
3
```

---

# Tupla com Diferentes Tipos

```python
dados = ("Ana", 20, 8.5, True)

print(dados)
```

Saída:

```python
('Ana', 20, 8.5, True)
```

---

# Tupla com Valores Repetidos

```python
numeros = (1, 2, 2, 3, 3, 3)

print(numeros)
```

Saída:

```python
(1, 2, 2, 3, 3, 3)
```

---

# Percorrendo uma Tupla

```python
frutas = ("maçã", "banana", "uva")

for fruta in frutas:
    print(fruta)
```

Saída:

```python
maçã
banana
uva
```

---

# Verificando se um Elemento Existe

```python
frutas = ("maçã", "banana", "uva")

print("banana" in frutas)
print("laranja" in frutas)
```

Saída:

```python
True
False
```

---

# Fatiamento (Slice)

```python
numeros = (10, 20, 30, 40, 50)

print(numeros[1:4])
```

Saída:

```python
(20, 30, 40)
```

---

# Tuplas são Imutáveis

```python
frutas = ("maçã", "banana", "uva")

frutas[0] = "laranja"
```

Saída:

```python
TypeError: 'tuple' object does not support item assignment
```

Não é possível alterar elementos diretamente.

---

# Convertendo Tupla para Lista

Quando precisamos modificar os dados.

```python
frutas = ("maçã", "banana", "uva")

lista = list(frutas)

lista.append("laranja")

print(lista)
```

Saída:

```python
['maçã', 'banana', 'uva', 'laranja']
```

---

# Convertendo Lista para Tupla

```python
lista = ["maçã", "banana", "uva"]

tupla = tuple(lista)

print(tupla)
```

Saída:

```python
('maçã', 'banana', 'uva')
```

---

# Contando Ocorrências

Método `count()`.

```python
numeros = (1, 2, 2, 3, 2, 4)

print(numeros.count(2))
```

Saída:

```python
3
```

---

# Encontrando a Posição de um Elemento

Método `index()`.

```python
frutas = ("maçã", "banana", "uva")

print(frutas.index("banana"))
```

Saída:

```python
1
```

---

# Desempacotamento (Unpacking)

```python
dados = ("Ana", 20, "Blumenau")

nome, idade, cidade = dados

print(nome)
print(idade)
print(cidade)
```

Saída:

```python
Ana
20
Blumenau
```

---

# Exemplo Prático – Cadastro Simples

# Exemplo Prático – Coordenadas

Muito utilizado em jogos e gráficos.

```python
ponto = (10, 25)

x, y = ponto

print("X =", x)
print("Y =", y)
```

Saída:

```python
X = 10
Y = 25
```

# Tupla com Tuplas

```python
alunos = (
    ("Ana", 8.5),
    ("Bruno", 7.0),
    ("Carlos", 9.0)
)

for nome, nota in alunos:
    print(nome, "-", nota)
```

Saída:

```python
Ana - 8.5
Bruno - 7.0
Carlos - 9.0
```

# Comparação: Lista x Tupla

| Lista                      | Tupla                  |
| -------------------------- | ---------------------- |
| `[]`                       | `()`                   |
| Mutável                    | Imutável               |
| Pode adicionar/remover     | Não pode alterar       |
| Mais flexível              | Mais segura            |
| Ideal para dados variáveis | Ideal para dados fixos |

### Exemplo

**Lista**

```python
alunos = ["Ana", "Bruno", "Carlos"]
```

Podemos adicionar novos elementos.

---

**Tupla**

```python
alunos = ("Ana", "Bruno", "Carlos")
```

Os dados permanecem fixos.

---

# Quando usar Tuplas?

✅ Coordenadas (x, y)

✅ Datas (dia, mês, ano)

✅ Dados que não devem ser alterados

✅ Retorno de múltiplos valores de funções

✅ Configurações fixas do sistema

---

# Resumo

| Operação             | Comando         |
| -------------------- | --------------- |
| Criar                | `()`            |
| Acessar              | `tupla[indice]` |
| Contar ocorrências   | `count()`       |
| Localizar posição    | `index()`       |
| Tamanho              | `len()`         |
| Verificar existência | `in`            |
| Converter para lista | `list()`        |
| Converter para tupla | `tuple()`       |

### Regra prática

> **Lista** → quando os dados podem mudar.
> **Tupla** → quando os dados devem permanecer os mesmos.

Exemplo: uma lista de alunos matriculados pode mudar; já uma coordenada `(10, 25)` ou uma data `(14, 6, 2026)` normalmente permanece fixa.
